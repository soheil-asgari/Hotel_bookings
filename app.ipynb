```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"./data/Hotel Bookings.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

We draw a graph to better understand the data distribution

```python
df.hist(bins=40, figsize=(20, 14))
plt.show()
```

Here we see the number of empty values â€‹â€‹in each column

```python
df.isnull().sum()
```

And for the company column, because there are many data without values, we delete it

```python
df = df.drop("company", axis=1)
```

We use pandas to fill in the non values

```python
df["children"].fillna(method="pad", inplace=True)
df["country"].fillna(method="pad", inplace=True)
df["agent"].fillna(method="bfill", inplace=True)
```

Duplicate data is deleted

```python
df = df.drop_duplicates()
```

From here on, we will call the data that we worked with the name of Hotel_bookings

```python
Hotel_bookings = df
```

There  are  two  types  of  hotels  in  the  dataset,  which  type  of  hotels 
booked most? Plot a pie chart to show the results.

We check the status of reservations

```python
Hotel_bookings["reservation_status"].unique()
```

We can see which hotel had more requests for reservations

```python
most_reserv = Hotel_bookings.groupby("hotel")["reservation_status"].count()
most_reserv
```

Enter the required libraries to draw the graph

```python
sns.set_style("white")
plt.rcParams["font.size"] = 9.0
colors = ["#9c9c9c", "#4e4e4e"]
plt.figure(figsize=(12, 8))
patches, texts, autotexts = plt.pie(
    data=Hotel_bookings,
    x=most_reserv,
    labels=most_reserv.index,
    autopct="%1.1f%%",
    colors=colors,
)
texts[0].set_fontsize(10)
plt.setp(autotexts, **{"color": "white", "weight": "bold", "fontsize": 12.5})
plt.setp(texts, **{"color": "gray", "weight": "bold", "fontsize": 10.5})
plt.title(
    "Hotels booked most",
    fontdict={"color": "#4e4e4e", "weight": "bold", "fontsize": 12.5},
    loc="left",
)
plt.show()
```
